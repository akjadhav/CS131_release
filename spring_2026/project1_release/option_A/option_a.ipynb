## Part 1: Convolutions
### 1.1 Commutative Property (5 points)
Recall that the convolution of an image $f:\mathbb{R}^2\rightarrow \mathbb{R}$ and a kernel $h:\mathbb{R}^2\rightarrow\mathbb{R}$ is defined as follows:
$$(f*h)[m,n]=\sum_{i=-\infty}^\infty\sum_{j=-\infty}^\infty f[i,j]\cdot h[m-i,n-j]$$

Or equivalently,
\begin{align}
(f*h)[m,n] &= \sum_{i=-\infty}^\infty\sum_{j=-\infty}^\infty h[i,j]\cdot f[m-i,n-j]\\
&= (h*f)[m,n]
\end{align}

Show that this is true (i.e. prove that the convolution operator is commutative: $f*h = h*f$).

**Your Answer:**

We start from the definition of convolution:

$$(f*h)[m,n]=\sum_{i=-\infty}^{\infty}\sum_{j=-\infty}^{\infty} f[i,j]\cdot h[m-i,n-j]$$

We perform the change of variables $i' = m - i$ and $j' = n - j$, so that $i = m - i'$ and $j = n - j'$. As $i$ ranges over all integers, so does $i'$ (and likewise for $j'$). Substituting:

$$
(f*h)[m,n] = \sum_{i'=-\infty}^{\infty}\sum_{j'=-\infty}^{\infty} f[m-i',\, n-j']\cdot h[i',\, j']
$$

By the definition of convolution (with $h$ as the first argument and $f$ as the second), this is exactly:

$$
= (h*f)[m,n]
$$

Therefore $f * h = h * f$, and convolution is commutative. $\blacksquare$

### 1.2 Shift Invariance (5 points)
Let $f$ be a function $\mathbb{R}^2\rightarrow\mathbb{R}$. Consider a system $f\xrightarrow{s}g$, where $g=(f*h)$ with some kernel $h:\mathbb{R}^2\rightarrow\mathbb{R}$. Also consider functions $f'[m,n] = f[m-m_0, n-n_0]$ and $g'[m,n] = g[m-m_0, n-n_0]$.  

Show that $S$ defined by any kernel $h$ is a shift invariant system by showing that $g' = (f'*h)$.

**Your Answer:**

We want to show that $(f'*h)[m,n] = g'[m,n]$.

By definition of convolution:

$$(f'*h)[m,n] = \sum_{i=-\infty}^{\infty}\sum_{j=-\infty}^{\infty} f'[m-i,\, n-j]\cdot h[i,j]$$

Substituting $f'[m-i, n-j] = f[(m-i)-m_0,\, (n-j)-n_0] = f[(m-m_0)-i,\, (n-n_0)-j]$:

$$= \sum_{i=-\infty}^{\infty}\sum_{j=-\infty}^{\infty} f[(m-m_0)-i,\, (n-n_0)-j]\cdot h[i,j]$$

By the definition of convolution, this is:

$$= (f*h)[m-m_0,\, n-n_0]$$

Since $g = f*h$, we have:

$$= g[m - m_0,\, n - n_0] = g'[m,n]$$

Therefore $(f'*h) = g'$, which proves that the system $S$ is shift invariant. $\blacksquare$


### 1.3 Linearity (10 points)

Recall that a system S is considered a linear system if and only if it satisfies the superposition property. In mathematical terms, a (function) S is a linear invariant system iff it satisfies:

$$
S\{\alpha f_1[n,m] + \beta f_2[n,m]\} = \alpha S\{f_1[n,m]\} + \beta S\{f_2[n,m]\}
$$

Let $f_1$ and $f_2$ be functions $\mathbb{R}^2\rightarrow\mathbb{R}$. Consider a system $f\xrightarrow{s}g$, where $g=(f*h)$ with some kernel $h:\mathbb{R}^2\rightarrow\mathbb{R}$.  

Prove that $S$ defined by any kernel $h$ is linear by showing that the superposition property holds.

**Your Answer:**

We need to show $S\{\alpha f_1 + \beta f_2\} = \alpha\, S\{f_1\} + \beta\, S\{f_2\}$, where $S\{f\} = f * h$.

By the definition of convolution:

$$S\{\alpha f_1 + \beta f_2\}[m,n] = \sum_{i}\sum_{j} \big(\alpha f_1[m-i,\, n-j] + \beta f_2[m-i,\, n-j]\big)\cdot h[i,j]$$

Distributing the multiplication over the sum:

$$= \sum_{i}\sum_{j} \alpha\, f_1[m-i,\, n-j]\cdot h[i,j] + \sum_{i}\sum_{j} \beta\, f_2[m-i,\, n-j]\cdot h[i,j]$$

Pulling the constants $\alpha$ and $\beta$ out of the summations:

$$= \alpha \sum_{i}\sum_{j} f_1[m-i,\, n-j]\cdot h[i,j] + \beta \sum_{i}\sum_{j} f_2[m-i,\, n-j]\cdot h[i,j]$$

$$= \alpha\,(f_1 * h)[m,n] + \beta\,(f_2 * h)[m,n]$$

$$= \alpha\, S\{f_1\}[m,n] + \beta\, S\{f_2\}[m,n]$$

Therefore the superposition property holds, and $S$ is a linear system. $\blacksquare$

#### Interpretation
How does the output of cross-correlation filter look? Explain what problems there might be with using a raw template as a filter.

**Your Answer:**

The cross-correlation output is brightest in regions of the image that are generally bright, regardless of whether the template is actually present there. This is because the raw template has all positive (non-zero mean) pixel values, so the dot product between the template and any bright region of the image will be large. The filter essentially responds to overall brightness rather than structural similarity to the template. As a result, raw cross-correlation is unreliable for template matching — it tends to produce false positives in bright areas and miss the template if it appears in a darker region of the image.

A solution is to normalize the pixels of the image and template at every step before comparing them. This is called **normalized cross-correlation**.

The mathematical definition for normalized cross-correlation of $f$ and template $g$ is:
$$(g \star f)[m,n]=\sum_{i,j} \frac{g[i, j]-\overline{g}}{\sigma_g} \cdot \frac{f[m + i, n + j]-\overline{f_{m,n}}}{\sigma_{f_{m,n}}}$$

where:
- $f_{m,n}$ is the patch image at position $(m,n)$
- $\overline{f_{m,n}}$ is the mean of the patch image $f_{m,n}$
- $\sigma_{f_{m,n}}$ is the standard deviation of the patch image $f_{m,n}$ 
- $\overline{g}$ is the mean of the template $g$
- $\sigma_g$ is the standard deviation of the template $g$

Implement **`normalized_cross_correlation`** function in **`filters.py`** and run the code below.

```python
from filters import normalized_cross_correlation

# Perform normalized cross-correlation between the image and the template
out = normalized_cross_correlation(img_gray, temp_gray)

# Find the location with maximum similarity
y, x = np.unravel_index(out.argmax(), out.shape)

# Display image
plt.imshow(img)
plt.title('Result (red marker on the detected location)')
plt.axis('off')

# Draw marker at detcted location
plt.plot(x, y, 'rx', ms=25, mew=5)
plt.show()
```

## Part 3: Separable Filters

### 3.1 Theory (10 points)
Consider an $M_1\times{N_1}$ image $I$ and an $M_2\times{N_2}$ filter $F$. A filter $F$ is **separable** if it can be written as a product of two 1D filters: $F=F_1F_2$.

For example,
$$F=
\begin{bmatrix}
1 & -1 \\
1 & -1
\end{bmatrix}
$$
can be written as a matrix product of
$$F_1=
\begin{bmatrix}
1  \\
1
\end{bmatrix},
F_2=
\begin{bmatrix}
1 & -1
\end{bmatrix}
$$
Therefore $F$ is a separable filter.

Prove that for any separable filter $F=F_1F_2$,
$$I*F=(I*F_1)*F_2$$
where $*$ is the convolution operation.

**Your Answer:**

Since $F$ is separable, we can write $F[i,j] = F_1[i]\cdot F_2[j]$, where $F_1$ is an $M_2 \times 1$ column vector and $F_2$ is a $1 \times N_2$ row vector.

Starting from the definition of 2D convolution:

$$(I * F)[m,n] = \sum_{i}\sum_{j} I[m-i,\, n-j]\cdot F[i,j]$$

Substituting $F[i,j] = F_1[i]\cdot F_2[j]$:

$$= \sum_{i}\sum_{j} I[m-i,\, n-j]\cdot F_1[i]\cdot F_2[j]$$

We can rearrange the sums since $F_2[j]$ does not depend on $i$:

$$= \sum_{j} F_2[j] \left(\sum_{i} I[m-i,\, n-j]\cdot F_1[i]\right)$$

The inner sum is exactly the convolution of $I$ with $F_1$ evaluated at $(m, n-j)$:

$$= \sum_{j} F_2[j]\cdot (I * F_1)[m,\, n-j]$$

This outer sum is the convolution of $(I * F_1)$ with $F_2$ evaluated at $(m, n)$:

$$= ((I * F_1) * F_2)[m,n]$$

Therefore $I * F = (I * F_1) * F_2$. $\blacksquare$

### 3.2 Complexity comparison (10 points)
Consider an $M_1\times{N_1}$ image $I$ and an $M_2\times{N_2}$ filter $F$ that is separable (i.e. $F=F_1F_2$).

(i) How many multiplication operations do you need to do a direct 2D convolution (i.e. $I*F$)?<br>
(ii) How many multiplication operations do you need to do 1D convolutions on rows and columns (i.e. $(I*F_1)*F_2$)?<br>
(iii) Use Big-O notation, written with respet to the dimensions $M_1$, $N_1$, $M_2$, and $N_2$, to argue which one is more efficient in general: direct 2D convolution or two successive 1D convolutions?

**Your Answer:**

**(i)** For direct 2D convolution $I * F$: at each of the $M_1 \cdot N_1$ output pixels, we multiply every element of the $M_2 \times N_2$ kernel with the corresponding image pixel. This requires:

$$M_1 \cdot N_1 \cdot M_2 \cdot N_2 \text{ multiplications}$$

**(ii)** For two successive 1D convolutions $(I * F_1) * F_2$:
- Convolving with $F_1$ ($M_2 \times 1$): at each of the $M_1 \cdot N_1$ pixels, we perform $M_2$ multiplications, giving $M_1 \cdot N_1 \cdot M_2$ multiplications.
- Convolving with $F_2$ ($1 \times N_2$): at each of the $M_1 \cdot N_1$ pixels, we perform $N_2$ multiplications, giving $M_1 \cdot N_1 \cdot N_2$ multiplications.
- Total:

$$M_1 \cdot N_1 \cdot (M_2 + N_2) \text{ multiplications}$$

**(iii)** In Big-O notation:
- Direct 2D convolution: $O(M_1 N_1 M_2 N_2)$
- Separable 1D convolutions: $O(M_1 N_1 (M_2 + N_2))$

Since $M_2 + N_2 \leq M_2 \cdot N_2$ for $M_2, N_2 \geq 2$, the separable approach is more efficient in general. The savings become more significant as the kernel size grows — for example, with a $k \times k$ kernel, separable convolution reduces complexity from $O(k^2)$ to $O(k)$ per pixel.
